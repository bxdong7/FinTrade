### DCA

```python
def dollar_cost_average(stock_data, investment_amount, frequency, investment_day, stock_exp_ratio, start_date, end_date):
    """
    This function is used to calculate and return the dollar cost average (DCA) for a given ETF and time frame.

    Args:
        stock_data : pandas dataframe
            This is the dataframe that contains all of the stocks being monitored for this project. It is a collection of 39 ETFs.
        investment_amount : int64
            Variable that represents the amount of money being invested in the stock.
        frequency : str
            This argument tells the function how often the trades will occur (i.e weekly, monthly or daily)
        investment_day : str
            This tells the function which day of the week the trades occur. (i.e Monday, Tuesday...)
        stock_exp_ratio : 
        start_date : dateTime
            This variable tells the function which date the trading will begin. And it is stored as a MM/DD/YYYY date format.
        end_date : dateTime
            This variable tells the function which date the trading will end. And it is stored as a MM/DD/YYYY date format.

    Returns:
        Total Investment : float64
            Returns the total amount of money invested.
        Final Value : float64
            Returns the final value of the investment.
        Annualized Return : float64
            This value represents the average of an investments earnings in a one year period.
        Annualized Roi : float64
            This value represents the rate of return for a given period less than one year but computes the value as if it were one year.
        Sharpe Ratio : float64
            This ratio is used to measure risk-relative returns of a trade.
        Longest Losing Streak : int64
            This value represents the longest number of trading days in a row that the function resulted in a loss.
    """
    df = stock_data
    investment_amount = int(investment_amount)
   
    # Create a calendar object for US holidays
    cal = USFederalHolidayCalendar()
    # Create a dictionary of holidays for faster lookup
    holiday_dict = holidays.US()
    
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
      
    # Define a function to get the next trading day, skipping weekends and holidays
    def get_next_trading_day(date):
        """
        This is a simple function that is used for finding the next possible trading day. 

        Args:
            date : str
                Variable that represents the day of the week.
        Returns:
            date : str
                This is the next available trading day according to the USFederalHolidayCalendar.
        """
        while date.weekday() >= 5 or date in holiday_dict or date in cal.holidays():
            date += BDay(1)
        return date

    # Convert investment_day to a weekday number (Monday=0, Sunday=6)
    #weekday = datetime.strptime(investment_day, '%A').weekday()
    #print('weekday', weekday)
    
    
    if(investment_day.lower() == 'monday'):
        weekday = 0
    elif(investment_day.lower() == 'tuesday'):
        weekday = 1
    elif(investment_day.lower() == 'wednesday'):
        weekday = 2
    elif(investment_day.lower() == 'thursday'):
        weekday = 3
    elif(investment_day.lower() == 'friday'):
        weekday = 4
    
    
    
    
    # Initialize variables
    total_investment = 0
    shares_bought = 0
    last_date = None
    longest_loss_streak = 0
    current_loss_streak = 0
    prev_price = None
    returns = []
    invest_dates = []
    invest_values = []
    close_for_sharpe = []
    # Loop through each investment date and buy shares
    date = start_date
    while date <= end_date:
        # Determine if investment_day falls on a weekend or holiday
        investment_day = datetime(date.year, date.month, date.day) + timedelta(days=(weekday - date.weekday()) % 7)
        #print('investment_day: ', investment_day)
        investment_day = get_next_trading_day(investment_day)
        #print('next trading investment_day: ', investment_day)
        while (investment_day not in df.index):
            investment_day += BDay(1)
            
        # Check if investment_day is within the start and end dates
        if start_date <= investment_day <= end_date:
            # Calculate the investment amount and buy shares at the closing price
            #print(investment_day)
            investment_amount_adjusted = min(investment_amount, total_investment + investment_amount)
            
            #shares_to_buy = investment_amount_adjusted / df.loc[investment_day, 'Close']
            #incorporated exp ratio
            shares_to_buy = (investment_amount * (1 - stock_exp_ratio/100)) / df.loc[investment_day, 'Close']

            shares_bought += shares_to_buy
            total_investment += investment_amount_adjusted

            # Track investment date and value
            invest_dates.append(investment_day)
            invest_values.append(investment_amount_adjusted)
    
            # Calculate daily returns
            if last_date:
                daily_return = (df.loc[investment_day, 'Close'] - prev_price) / prev_price
                returns.append(daily_return)

                # Track longest losing streak
                if daily_return < 0:
                    current_loss_streak += 1
                    longest_loss_streak = max(longest_loss_streak, current_loss_streak)
                else:
                    current_loss_streak = 0

            # Track the last date and closing price
            last_date = investment_day
            prev_price = df.loc[investment_day, 'Close']
            
        close_for_sharpe.append(df.loc[investment_day, 'Close'])
            
        # Increment the date by the specified frequency
        if frequency == 'daily':
            date += timedelta(days=1)
        elif frequency == 'weekly':
            date += timedelta(weeks=1)
        elif frequency == 'biweekly':
            date += timedelta(weeks=2)
        elif frequency == 'monthly':
            date = date.replace(day=1) + timedelta(days=31)
        elif frequency == 'bimonthly':
            date = date.replace(day=1) + timedelta(days=61)
        else:
            raise ValueError('Invalid frequency parameter')


    #total_investment = round(total_investment, 2)
    final_value = shares_bought * df.loc[get_next_trading_day(end_date), 'Close']
    annualized_return = ((final_value / total_investment) ** (1 / ((end_date - start_date).days / 365)))
    

    #stock_annualized_return = ((close_price*stock_shares_owned)/stock_total_investment) ** (1/num_years) 
    
    
    annualized_roi = (final_value - total_investment) / total_investment / ((end_date - start_date).days / 365)
    #sharpe_ratio = round(np.sqrt(252) * np.mean(returns) / np.std(returns), 4)
    
    daily_returns = df['Close'].pct_change()

    close_for_sharpe_df = pd.DataFrame(close_for_sharpe, columns=['Close'])
        
    #this works!!!!
    this_returns = close_for_sharpe_df['Close'].pct_change()
    sharpe_ratio = np.sqrt((end_date - start_date).days / 365) * this_returns.mean() / this_returns.std()
        

    
    losing_streaks = (daily_returns < 0).astype(int).groupby(pd.Grouper(freq='W')).sum()
    longest_losing_streak = losing_streaks.max()

    
    #output = {
     #   'Total investment': total_investment,
      #  'Final value': final_value,
       # 'Annualized return': annualized_return,
        #'Annualized ROI': annualized_roi,
        #'Sharpe ratio': sharpe_ratio,
        #'Longest losing streak (weeks)': longest_loss_streak/5 
        #'Longest losing streak (weeks)': longest_losing_streak 
   # }

    #return output
    return total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak 
```

### Rebalance

```python
#rebalance - DONE

def rebalance(input_etfs, weights, start_date, end_date, frequency):
    """
    This function preforms a rebalancing strategy on a list of ETFs that make up a portfolio.
    
    Args:
        input_etfs : list
            This is the list of ETFs that the portfolio is holding.
        weights : list
            This list represents the weights of each of the ETFs.
        start_date : dateTime
            This represents when the rebalancing will start. Stored as MM/DD/YYYY
        end_date : dateTime
            This represents when the rebalancing period is over. Stored in MM/DD/YYYY format.
        frequency : string
            This variable represents how often the etfs will be rebalanced.

    Returns:
        Rebalanced Returns : float64
            This value represents the returns of the portfolio after the rebalancing has been complete.
        Sharpe Ratio : float64
            The sharpe ratio is how you measure returns by taking risk into consideration.


    """
    
    prices = pd.DataFrame()
    rebalance_frequency = float(frequency)
    
    for etf in input_etfs:
        stock_data = etf_dict[etf]
        prices[etf] = stock_data['Close']
    
    # Select data for the rebalancing period
    prices = prices.loc[start_date:end_date]
   
    # Compute returns and annualized returns for each ETF
    returns = prices.pct_change().dropna()
    #print('returns ', returns)
    annualized_returns = (1 + returns.mean()) ** 252 - 1
    
    # Compute the current weights of each ETF in the portfolio
    current_weights = pd.Series(weights, index=input_etfs)
    
    # Compute the current value of the portfolio
    current_value = (current_weights * prices.iloc[-1]).sum()
    
    # Compute the target value of each ETF in the portfolio
    target_value = current_value * current_weights
    
    # Compute the required number of shares for each ETF
    share_counts = (target_value / prices.iloc[-1])
    
    # Compute the actual weights of each ETF in the rebalanced portfolio
    actual_weights = share_counts * prices.iloc[-1] / current_value
    
    # Compute the rebalanced portfolio's annualized return
    periods_per_year = 0
    if rebalance_frequency == 1:
        # Rebalance every year
        rebalanced_returns = annualized_returns.mean()
        periods_per_year = 1
        
    elif rebalance_frequency == 2:
        # Rebalance every 6 months
        rebalanced_returns = ((1 + annualized_returns).rolling(window=2).apply(np.prod, raw=True) ** (0.5) - 1).mean()
        periods_per_year = 2
    
    # Compute the Sharpe Ratio
    # Compute the daily returns for the rebalanced portfolio
    daily_returns = (actual_weights * returns).sum(axis=1)
    #print('daily_returns ', daily_returns)
    # Compute the mean and standard deviation of the daily returns
    mean_return = daily_returns.mean()
    std_dev = daily_returns.std()
    
    # Compute the Sharpe Ratio assuming a risk-free rate of 0
    #sharpe_ratio = mean_return / std_dev
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    sharpe_ratio = np.sqrt(periods_per_year)*np.sqrt(((end_date - start_date).days / 365))*mean_return / std_dev
    

    
    
    return rebalanced_returns, sharpe_ratio

    
```

```python
def Buy_Sell_Strategy(stocks, pct_changes, buy_ins, start_date, end_date):

    """
    This function preforms a buy-sell strategy on a given list of stocks. 
    
    Args:
        stocks : list
            Represents a list of stocks that are going to constitute the portfolio.
        pct_changes : float64
            Represents the percent change that will cause the function to sell the stock.
        buy_ins : float64
            Represents the percent change that will cause the function to buy another share of the stock(s).
        start_date : dateTime
            Date that represents when this strategy will start. Stored as MM/DD/YYYY
        end_date : dateTime
            Date that represents when the strategy will end. Stored as MM/DD/YYYY

    Returns: 
        portfolio_stats : dataframe
            This dataframe that gets returned contains information about all of the stocks that were either bought or sold during this strategy.
            The dataframe contains: sharpe ratio, total investment, annualized returns, and total shares owned for each stock.
    """
    # Load the stock data from the CSV files
    buy_stock_data = {}
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    for stock in stocks:
        stock_name = stock.split('.')[0]
        #print(stock_file)
        #stock_data[stock_name] = pd.read_csv(f"IndexFundsData/{stock_name}.csv")
        
        # Convert the 'date' column to a datetime object
        # stock_data[stock_name]['Date'] = pd.to_datetime(stock_data[stock_name]['Date'])
        #print(stock_name)
        buy_stock_data[stock_name] = etf_dict[stock].reset_index()
        #print(stock_data[stock_name])
        buy_stock_data[stock_name]['Date'] = pd.to_datetime(buy_stock_data[stock_name]['Date'])
        
        # Filter the stock data based on the specified start and end dates
        buy_stock_data[stock_name] = buy_stock_data[stock_name][(buy_stock_data[stock_name]['Date'] >= start_date) & 
                                                        (buy_stock_data[stock_name]['Date'] <= end_date)]
        
        # Calculate the daily returns
        buy_stock_data[stock_name]['returns'] = buy_stock_data[stock_name]['Close'].pct_change()

    # Initialize the portfolio
    balance = 0
    total_investment = 0
    shares_owned = {stock_name: 0 for stock_name in buy_stock_data}
    last_trade_date = {stock_name: None for stock_name in buy_stock_data}
    portfolio_stats = {}
    
    # Iterate through each stock
    for stock_name, stock_df in buy_stock_data.items():
        
        stock_pct_change = (pct_changes[stock_name]/100)
        stock_buy_in = buy_ins[stock_name]
        
        # Initialize stock-specific portfolio variables
        stock_balance = 0
        stock_total_investment = 0
        stock_shares_owned = 0
        stock_last_trade_date = None
        
        close_for_sharpe = []
        
        # Iterate through the daily returns for this stock
        for i, row in stock_df.iterrows():
            date = row['Date']
            returns = row['returns']
            close_price = row['Close']

            # Check if we need to make a trade
            if stock_last_trade_date is not None and (date - stock_last_trade_date).days <= 1:
                continue  

            if (stock_pct_change < 0 and returns <= stock_pct_change) or (stock_pct_change > 0 and returns >= stock_pct_change):
                # Buy shares
                #shares_to_buy = buy_in
                shares_to_buy = stock_buy_in
                cost = shares_to_buy * close_price
                stock_balance -= cost
                stock_total_investment += cost
                stock_shares_owned += shares_to_buy
                stock_last_trade_date = date
                
                
                close_for_sharpe.append(close_price)
        # Calculate stock-specific portfolio statistics
        num_years = (stock_df['Date'].iloc[-1] - stock_df['Date'].iloc[0]).days / 365.25
        #stock_sharpe_ratio = np.sqrt(num_years) * stock_df['returns'].mean() / stock_df['returns'].std()
        

        
        
        close_for_sharpe_df = pd.DataFrame(close_for_sharpe, columns=['Close'])
        

        this_returns = close_for_sharpe_df['Close'].pct_change()
        stock_sharpe_ratio = np.sqrt(num_years) * this_returns.mean() / this_returns.std()
        

        
        stock_annualized_return = ((close_price*stock_shares_owned)/stock_total_investment) ** (1/num_years) 
        # Store the stock-specific portfolio statistics
        portfolio_stats[stock_name] = {'balance': stock_balance,
                                       'total_investment': stock_total_investment,
                                       'sharpe_ratio': stock_sharpe_ratio,
                                       'annualized_return': stock_annualized_return,
                                       "stock_shares_owned": stock_shares_owned}
        
        # Update the overall portfolio variables
        balance += stock_balance
        total_investment += stock_total_investment
        for stock_name in shares_owned:
            shares_owned[stock_name] += stock_shares_owned if stock_name == stock_name else 0

    # Calculate overall portfolio statistics
    combined_df = pd.concat(buy_stock_data.values(), axis=0)
    combined_shares_owned = shares_owned[list(shares_owned.keys())[0]]
    
    num_years = (end_date - start_date).days / 365.25
    #combined_sharpe_ratio = np.sqrt(num_years) * combined_df['returns'].mean() / combined_df['returns'].std()
    
    #same results
    #log_return = np.log(combined_df['Close']/combined_df['Close'].shift())
    #combined_sharpe_ratio = log_return.mean()/log_return.std()
    
    #----
    stock_sharpe_ratios = [v['sharpe_ratio'] for k, v in portfolio_stats.items()]
    #stock_sharpe_ratios = portfolio_stats[]['sharpe_ratio']
    stock_market_values = [v['total_investment'] for k, v in portfolio_stats.items()]
    stock_weights = [mv / total_investment for mv in stock_market_values]
    portfolio_sharpe_ratio = np.dot(stock_sharpe_ratios, stock_weights)
    #print("-----portfolio_sharpe_ratio: ", portfolio_sharpe_ratio)
    #--------
    #combined_annualized_return = (np.power(combined_df['Close'].iloc[-1] / combined_df['Close'].iloc[0], 1/num_years) - 1)
    combined_annualized_return = ((combined_df['Close'].iloc[-1]*combined_shares_owned)/total_investment) ** (1/num_years)
    
    # Store the overall portfolio statistics
    portfolio_stats['Combined'] = {'balance': balance,
                                   'total_investment': total_investment,
                                   'sharpe_ratio': portfolio_sharpe_ratio,
                                   'annualized_return': combined_annualized_return,
                                   'shares_owned': combined_shares_owned}

    return portfolio_stats
```

```python
def dca_portfolio(input_etfs, start_date, end_date):

    """
    This function preforms a dollar cost average strategy on a provided list of ETFs. 
    
    Args:
        input_etfs : list
            A list of ETFs that the strategy will invest in.
        start date : dateTime
            When the strategy will begin.
        end_date : dateTime
            When the stragy will end.

    Returns:
        dca_portfolio : dataFrame
            This dataframe contains the columns listed directly below \/, this function also prints out total investment, final value, annualized returns, annualized ROI, 
            sharpe ratio and the longest losing streak in weeks.
    """

    dca_portfolio = pd.DataFrame(columns=['Stock', 'expense_ratio','investment_amount', 'frequency', 'investment_day', 'total_investment', 'final_value', 'annualized_return',
                                         'annualized_roi', 'sharpe_ratio', 'longest_losing_streak'])
    
      
    for i in range(len(input_etfs)):
        
        if(start_date == 'inception'):
            s_date = stock_data_df.loc[stock_data_df['Stock'] == input_etfs[i], 'Inception_Date'].iloc[0]
            print(s_date)
        else:
            s_date = start_date
        
        print("Stock: ", input_etfs[i])
        #print(start_date)
        stock_data = etf_dict[input_etfs[i]]
        #print(stock_data)
        stock_exp_ratio = stock_data_df.loc[stock_data_df['Stock'] == input_etfs[i], 'Expense_Ratio'].iloc[0]
        print("Expense ratio: ", stock_exp_ratio)
        #investment_amount = input("Enter Investment amount: \n")
        investment_amount = 1000
        frequency = input("Enter frequency (daily, weekly, biweekly, monthly): \n")
        investment_day = input("Enter day of the week to invest: \n") 

        total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak = dollar_cost_average(stock_data, investment_amount, frequency, investment_day, stock_exp_ratio, s_date, end_date)

        dca_portfolio.loc[len(dca_portfolio.index)] = [input_etfs[i],stock_exp_ratio, investment_amount, frequency,investment_day, total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak ]
        print(f'Total investment:', total_investment)
        print(f'Final Value: ', final_value)
        print(f'Annualized return: {annualized_return:}')
        print(f'Annualized ROI: {annualized_roi:}')
        print(f'Sharpe ratio: {sharpe_ratio:.2}')
        print(f'Longest losing streak (weeks): {longest_losing_streak}\n')
        
     
    
    return dca_portfolio
```

```python
def dca_portfolio_multi(input_etfs, frequency_list, days, start_date, end_date):
    """
    This function is used to preform a dollar cost average strategy on a portfolio that contains more than one stock.
    
    Args:
        input_etfs : list
            A list that contains each of the stocks that will make up the portfolio.
        frequency_list : list
            A list that contains how frequently a stock will get traded. If there are 3 ETFs listed there needs to be three frequencies listed.
        days : list
            This list determines which day of the week the trading will occur.  If there are 3 ETFs listed there needs to be three frequencies listed.
        start_date : dateTime
            Determines which date the trading will start.
        end_date : dateTime
            Determines which date the trading will end.
    Returns:
        dca_portfolio : dataFrame
            This dataframe contains the columns listed directly below \/, this function also prints out total investment, final value, annualized returns, annualized ROI, 
            sharpe ratio and the longest losing streak in weeks.
    """

    dca_portfolio = pd.DataFrame(columns=['Stock', 'expense_ratio','investment_amount', 'frequency', 'investment_day', 'total_investment', 'final_value', 'annualized_return',
                                         'annualized_roi', 'sharpe_ratio', 'longest_losing_streak'])
    
      
    for i in range(len(input_etfs)):
        
        if(start_date == 'inception'):
            s_date = stock_data_df.loc[stock_data_df['Stock'] == input_etfs[i], 'Inception_Date'].iloc[0]
            print(s_date)
        else:
            s_date = start_date
        
        print("Stock: ", input_etfs[i])
        #print(start_date)
        stock_data = etf_dict[input_etfs[i]]
        #print(stock_data)
        stock_exp_ratio = stock_data_df.loc[stock_data_df['Stock'] == input_etfs[i], 'Expense_Ratio'].iloc[0]
        print("Expense ratio: ", stock_exp_ratio)
        #investment_amount = input("Enter Investment amount: \n")
        investment_amount = 1000
        frequency = frequency_list[i]
        investment_day = days[i] 

        total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak = dollar_cost_average(stock_data, investment_amount, frequency, investment_day, stock_exp_ratio, s_date, end_date)

        dca_portfolio.loc[len(dca_portfolio.index)] = [input_etfs[i],stock_exp_ratio, investment_amount, frequency,investment_day, total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak ]
        print(f'Total investment:', total_investment)
        print(f'Final Value: ', final_value)
        print(f'Annualized return: {annualized_return:}')
        print(f'Annualized ROI: {annualized_roi:}')
        print(f'Sharpe ratio: {sharpe_ratio:.2}')
        print(f'Longest losing streak (weeks): {longest_losing_streak}\n')
        
     
    
    return dca_portfolio
```

```python
#FINAL rebalance 
def rebalance_porfolio(input_etfs, start_date, end_date):
    """
    This function preforms a rebalancing stategy on a provided portfolio of stocks.
    
    Args:
        input_etfs : list
            A list containing the stock symbols for the ETFs that the user wants to trade.
        start_date : dateTime
            The date when the rebalancing will begin.
        end_date : dateTime
            The date when the rebalancing strategy will stop.

    Returns:
        etfs_rebalance : DataFrame
            This DataFrame contains the information about the stocks and their corresponding rebalancing weights.
        annualized_return : float64
            This value represents the annualized return of the portfolio.
        sharpe_ratio : float64
            This value represents the sharpe ratio for the portfolio.
    """
    weights = []

    while(True):
        weights=[]
        for e in input_etfs:
            w = input(f"Enter ratio for {e} : ")
            weights.append(float(w))
        total = sum(weights)
        if(total == 1):
            break
        print("Error: The total ratio must be 1.")   

    freq = input("Enter frequency of rebalance (1 for yearly, 2 biyearly): ")
    print('Ratios: ', weights)
    print('Rebalance frequency: ', freq)


    start_date = "2020-01-01"
    end_date = "2022-12-31"

    print(input_etfs)
    annualized_return, sharpe_ratio = rebalance(input_etfs, weights, start_date, end_date, freq)


    print(f"YEARLY: The annualized return of the rebalanced portfolio is {annualized_return:.2%}")
    print("Sharpe ratio: ", sharpe_ratio)

    
    e_rb = {'Stock': input_etfs, 'Ratio': weights}
    etfs_rebalance = pd.DataFrame(data=e_rb)
    etfs_rebalance 
    
    return etfs_rebalance, annualized_return, sharpe_ratio 
```

```python
#FINAL buy sell trade_rules

#start_date = '2020-01-01'
#end_date = '2022-12-31'

def buy_sell_portfolio(input_etfs, start_date, end_date):
    """

    This function preforms a buy-sell strategy on a provided list of ETFs. 
    
    Args: 
        input_etfs : list
            A list of the ETFs that the user is interested in trading.
        start_date : dateTime
            The date when the strategy will begin.
        end_date : dateTime
            The date at which the strategy will stop.

    Returns:
        Buy_in_out_portfolio : dataFrame
            This dataFrame contains the stats from the buy-sell strategy to see how well it preformed based on the stocks chosen.
        overall_df : dataFrame
            This is the dataFrame that contains overall statistics for the portfolio.
    """

    pct_changes = {}
    buy_ins = {}

    for e in input_etfs:
        p = input(f"Enter percent change for {e}: ")
        pct_changes[e] = float(p)
        b = input(f"Enter amount of shares for {e}: ")

        buy_ins[e] = int(b)


    # Calculate the portfolio statistics
    portfolio_stats = Buy_Sell_Strategy(input_etfs, pct_changes, buy_ins, start_date, end_date)
    print(input_etfs)
    print(pct_changes)
    print(buy_ins)
    print(" ")
    # Print the overall portfolio statistics
    print("Overall Portfolio Statistics:")
    for key, value in portfolio_stats['Combined'].items():
        print(f"{key}: {value:.2f}")

    # Print the individual stock statistics
    Buy_in_out_portfolio = pd.DataFrame(columns=['Stock', 'Percent_change', 'Shares', 'Balance', 'Total_investment', 'stock_shares_owned', 'Annualized_return', 'Sharpe_ratio'])

    print("\nIndividual Stock Statistics:")
    for stock_name, stats in portfolio_stats.items():
        if stock_name != 'Combined':
            print(f"{stock_name}:")
            for key, value in stats.items():
                print(f"  {key}: {value:.2f}")


    for x in input_etfs:
        Buy_in_out_portfolio.loc[len(Buy_in_out_portfolio.index)] = [x,pct_changes[x], buy_ins[x], 
                                                                     portfolio_stats[x]['balance'], 
                                                                     portfolio_stats[x]['total_investment'],
                                                                     portfolio_stats[x]['stock_shares_owned'],
                                                                     portfolio_stats[x]['annualized_return'],
                                                                     portfolio_stats[x]['sharpe_ratio']]
  
    combined_dict = portfolio_stats['Combined']
    overall_df = pd.DataFrame.from_dict(combined_dict, orient='index', columns=['Value'])
    overall_df.index.name = 'Overall Metrics'
    return Buy_in_out_portfolio, overall_df
```

```python
#start = "2017-01-01"
#end = "2022-12-31"
#investment_amount = 1000
#frequency = "weekly"
#investment_day =  "Monday"

def single_dca_test(etfs, start, end, investment_amount, frequency, investment_day): 
    """
    This function preforms a single dollar cost average strategy that uses the a list of ETFs provided by the user.

    Args:
        etfs : list
            This is the list of etfs that the user is interested in buying.
        start : dateTime
            This date represents when this strategy will begin.
        end : dateTime
            This date represents when the strategy will stop.
        investment_amount : int64
            This value represents the amount of money that will initially be invested in the listed ETFs
        frequency : string
            This value represents how often this strategy will take place. i.e Weekly will mean that the trades will occur once a week on the investment_day value.
        investment_day : string
            This value represents which day of the week the first trade will take place.

    Returns:
        dca_all_stocks : dataFrame
            This function returns a dataframe that contains the columns listed right below \/. This dataframe will give the end-user a roundabout idea of how well this strategy preformned.
    """

    dca_all_stocks= pd.DataFrame(columns=['Stock', 'expense_ratio','investment_amount', 'frequency', 'investment_day', 'total_investment', 'final_value', 'annualized_return',
                                             'annualized_roi', 'sharpe_ratio', 'longest_losing_streak'])

    for i in range(len(etfs)):
        stock_data = etf_dict[etfs[i]]
        stock_exp_ratio = stock_data_df.loc[stock_data_df['Stock'] == etfs[i], 'Expense_Ratio'].iloc[0]

        total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak = dollar_cost_average(stock_data, investment_amount, frequency, investment_day, stock_exp_ratio, start, end)

        dca_all_stocks.loc[len(dca_all_stocks.index)] = [etfs[i],stock_exp_ratio, investment_amount, frequency,investment_day, total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak ]

    print(f" {start} - {end}")    
    return dca_all_stocks
```

```python
def Single_dca_inception_date(etfs, frequency, investment_day, end_date):
    """
    This function will preform single dollar cost average strategy for the ETFs that the user wants from the ETFs inception until a specified end date.
    
    Args:
    etfs : list
        This is a list of ETFs that the user wants to preform this function on.
    frequency : string
        This string represents how frequent the investments will happen.
    investment_day : string
        This string represents which day of the week the trades will occur.
    end_date : dateTime
        This date represents when the strategy will stop.

    Returns: 
    dca_all_stocks : dataFrame
        This dataFrame contains information that can give the user an overview of how well the strategy preformed. The columns for this dataFrame can be found righr below.
    """
    
    
    dca_all_stocks= pd.DataFrame(columns=['Stock','inception_date','expense_ratio','investment_amount', 'frequency', 'investment_day', 'total_investment', 'final_value', 'annualized_return',
                                             'annualized_roi', 'sharpe_ratio', 'longest_losing_streak'])

    for i in range(len(etfs)):
        stock_data = etf_dict[etfs[i]]
        stock_exp_ratio = stock_data_df.loc[stock_data_df['Stock'] == etfs[i], 'Expense_Ratio'].iloc[0]
        
        incep_date = stock_data_df.loc[stock_data_df['Stock'] == etfs[i], 'Inception_Date'].iloc[0]
        
        total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak = dollar_cost_average(stock_data, investment_amount, frequency, investment_day, stock_exp_ratio, incep_date, end)

        dca_all_stocks.loc[len(dca_all_stocks.index)] = [etfs[i],incep_date, stock_exp_ratio, investment_amount, frequency,investment_day, total_investment, final_value, annualized_return, annualized_roi, sharpe_ratio, longest_losing_streak]

    #print(f" {start} - {end}")    
    return dca_all_stocks
    
```

```python
def get_multi_dca_overall_stats(df, start_date, end_date):
    """
    This function calculates the overal stats for the multi DCA investing strategy.

    Args:
        df : dataFrame
            This dataFrame contains the statistics for each stock invested in the multi-DCA strategy
        start_date : dateTime
            This is the date which the strategy will begin.
        end_date : dateTime
            This is the date which the strategy will end.

    Returns:
        overall_annualized_return : float64
            This value represents the overall annualized returns from this strategy.
        overall_sharpe_ratio : float64
            This value represents the overall sharpe ratio for this portfolio and its returns.

    """


    overall_final_value = df['final_value'].sum()
    overall_total_investment = df['total_investment'].sum()
    this_start = ""
    incep_date_list = []
    this_end = pd.to_datetime(end_date)
      
    if(start_date == 'inception'):
        for i in df['Stock']:
            
            x = stock_data_df.loc[stock_data_df['Stock'] == i, 'Inception_Date'].iloc[0]
            incep_date_list.append(x)
        this_start = pd.to_datetime(incep_date_list).min()
    else:
        this_start = pd.to_datetime(start_date)
        

    overall_annualized_return = (overall_final_value/overall_total_investment) ** (1/((this_end - this_start).days / 365))
    print('overall_annualized_return: ', overall_annualized_return)


    overall_sharpe_ratio = ((df['total_investment'] / overall_total_investment) * df['sharpe_ratio']).sum()
    print('overall_sharpe_ratio: ', overall_sharpe_ratio)
    
    return overall_annualized_return, overall_sharpe_ratio
```

```python
import pandas as pd
from datetime import datetime, timedelta

import pandas as pd



def rebalance2(start_date, end_date, frequency, portfolio_df):
    """
    This function is used to rebalance a portfolio using a given frequency. 

    Args:
        start_date : dateTime
            This date represents when the rebalancing strategy will start.
        end_date : dateTime
            This date represents when the rebalancing strategy will end.
        frequency : int64
            This string represents how frequently the portfolio will get rebalanced.
        portfolio_df : dataFrame
            This dataFrame contains information about the portfolio that is going to get rebalanced.

    Returns: 
        annualized_returns : float64
            This value represents the annualized returns for the portfolio using this rebalancing strategy.
        sharpe_ratio : float64
            This value represents the average sharpe ratio for this portfolio and its returns. 
    """
   
    # Calculate the number of months in the rebalancing period
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)
    months = (end.year - start.year) * 12 + (end.month - start.month) + 1
    
    # Determine the number of rebalancing periods based on the frequency
    if frequency == 1:
        periods = months // 12
        months_per_period = 12
        years = ((end - start).days / 365)
    elif frequency == 2:
        periods = months // 6
        months_per_period  = 6
        years = ((end - start).days / 365)/2
    else:
        raise ValueError("Invalid frequency specified. Must be 'once a year' or 'twice a year'.")
    
    # Calculate the weight of each stock in the portfolio
    total_investment = portfolio_df["total_investment"].sum()
    portfolio_df["weight"] = portfolio_df["total_investment"] / total_investment
    
    
    #print(periods)
   
    
    # Rebalance the portfolio for each period
    for i in range(periods):
        # Determine the start and end dates for the current rebalancing period
        period_start_date = start + pd.DateOffset(months=i*months_per_period)
        #print('period_start_date: ', period_start_date)
        period_end_date = period_start_date + pd.DateOffset(months=months_per_period) - pd.DateOffset(days=1)
        #print('period_end_date ', period_end_date)
        # Calculate the returns for each stock in the portfolio for the current period
        
        for j in range(len(portfolio_df)):
            stock = portfolio_df.loc[j, "Stock"]
            
            #period_start_date = get_next_trading_day(period_start_date)
            #print('next trading investment_day: ', investment_day)
            while (period_start_date not in etf_dict[stock].index):
                period_start_date += BDay(1)
            
            while (period_end_date not in etf_dict[stock].index):
                period_end_date -= BDay(1)
                
            #print('period_start_date: ', period_start_date)
            #print('period_end_date: ', period_end_date)
            start_price = etf_dict[stock].loc[period_start_date, "Close"]
            print('start_price ', start_price)
            end_price = etf_dict[stock].loc[period_end_date, "Close"]
            print('end_price ', end_price)
            returns = (end_price - start_price) / start_price
            print('returns ', returns) 
            #portfolio_df.loc[j, "final_value"] = portfolio_df.loc[j, "total_investment"] * (1 + returns)
            portfolio_df.loc[j, "new_final_value"] = (portfolio_df.loc[j, "total_investment"]/years) * (1 + returns)
            print('portfolio_df.loc[j, "new_final_value"] ', portfolio_df.loc[j, "new_final_value"])
        
        # Calculate the new investment amount for each stock based on the new weights
        total_value = portfolio_df["final_value"].sum()
        print('------total_value ', total_value)
        for j in range(len(portfolio_df)):
            weight = portfolio_df.loc[j, "weight"]
            print('weight: ', weight)
            portfolio_df.loc[j, "investment_amount"] = total_value * weight
        
        # Calculate the total investment in the portfolio
        portfolio_df["total_investment"] = portfolio_df["investment_amount"] * (1 + portfolio_df["sharpe_ratio"])
    
    # Calculate the overall annualized return and Sharpe ratio after rebalancing
    final_value = portfolio_df["total_investment"].sum()
    initial_value = portfolio_df["total_investment"].sum() / (1 + portfolio_df["annualized_return"])**periods
    total_returns = (final_value - initial_value) / initial_value
    annualized_returns = (1 + total_returns)**(12/months) - 1
    sharpe_ratio = portfolio_df["sharpe_ratio"].mean()
    
    
    
    return portfolio_df, annualized_returns, sharpe_ratio 

```

```python


def rebalance3(start_date, end_date, frequency, portfolio):
    """
    This function is used to rebalance a portfolio using a given frequency. 

    Args:
        start_date : dateTime
            This date represents when the rebalancing strategy will start.
        end_date : dateTime
            This date represents when the rebalancing strategy will end.
        frequency : string
            This string represents how frequently the portfolio will get rebalanced.
        portfolio : dataFrame
            This dataFrame contains information about the portfolio that is going to get rebalanced.

    Returns: 
        overall_return : float64
            This value represents the overall return for the portfolio using the rebalancing strategy.
        overall_sharpe_ratio : float64
            This value represents the sharpe ratio for this portfolio and its returns. 
    """
    # Step 1: Determine the rebalancing dates
    if frequency == 1:
        rebal_dates = pd.date_range(start=start_date, end=end_date, freq="AS")
    elif frequency == 2:
        rebal_dates = pd.date_range(start=start_date, end=end_date, freq="6MS")
    else:
        raise ValueError("Invalid frequency. Allowed values: 'once a year', 'twice a year'")

    # Step 2: Iterate over the rebalancing dates and rebalance the portfolio
    for date in rebal_dates:
        # Get the current portfolio weights
        weights = portfolio["investment_amount"] / portfolio["total_investment"]

        # Calculate the current value of the portfolio
        current_value = np.sum(portfolio["final_value"])

        # Calculate the desired value of the portfolio
        desired_value = np.sum(portfolio["investment_amount"])

        # Calculate the required adjustments to each stock
        adjustments = (desired_value * weights / etf_dict[portfolio["Stock"][0]]["Close"].loc[date]) - portfolio["investment_amount"]

        # Apply the adjustments to each stock
        for i in range(len(portfolio)):
            stock = portfolio["Stock"][i]
            investment_amount = portfolio["investment_amount"][i]
            portfolio["investment_amount"][i] += adjustments[i]
            portfolio["total_investment"][i] += adjustments[i]
            portfolio["final_value"][i] = etf_dict[stock]["Close"].loc[date] * portfolio["investment_amount"][i]
            portfolio["annualized_return"][i] = ((portfolio["final_value"][i] / current_value) ** (1 / (len(rebal_dates) + 1))) - 1

    # Step 3: Calculate the overall annualized return and Sharpe ratio
    overall_return = np.prod(1 + portfolio["annualized_return"]) - 1
    overall_sharpe_ratio = np.average(portfolio["sharpe_ratio"], weights=portfolio["investment_amount"])

    return overall_return, overall_sharpe_ratio

```

```python
    
def Single_buy_in_out_test(etfs, start_date, end_date, pct_c, shares):
    """
    This function preforms a single buy-in-out strategy for a porfolio of ETFs.

    Args:
        etfs : list
            This is the list of ETFs that the user wants to preofrm this strategy on. 
        start_date : dateTime
            This is the date for when the strategy will begin. 
        end_date : dateTime
            This is the date for when the strategy will end. 
        pct_c : float64
            This value represents the percent chance for when to buy or sell shares. 
        shares : int64
            This value represents the number of shares to buy.
    
    Returns:
        Buy_in_out_portfolio : dataFrame
            This dataFrame contains the annualized returns and the sharpe ratio for each of the stocks that the user input into this function.
    """
    pct_changes = {}
    buy_ins = {}

    for e in etfs:
        pct_changes[e] = pct_c
        buy_ins[e] = shares

    # Calculate the portfolio statistics
    portfolio_stats = Buy_Sell_Strategy(etfs, pct_changes, buy_ins, start_date, end_date)

    #return portfolio_stats
    
    Buy_in_out_portfolio = pd.DataFrame(columns=['Stock', 'Percent_change', 'Shares', 'Annualized_return', 'Sharpe_ratio'])

    for x in etfs:
        Buy_in_out_portfolio.loc[len(Buy_in_out_portfolio.index)] = [x,pct_changes[x], buy_ins[x], 
                                                                     portfolio_stats[x]['annualized_return'],
                                                                     portfolio_stats[x]['sharpe_ratio']]
        
    return Buy_in_out_portfolio
```

```python
#stock_data_df
#etfs
#end_date = '2022-12-31'

def Single_buy_in_inception_date(etfs, pct_c, shares, end_date):
    """
    This function preforms a single buy-in-out strategy for a porfolio of ETFs since the inception of those ETFs.

    Args:
        etfs : list
            This is the list of ETFs that the user wants to preofrm this strategy on. 
        start_date : dateTime
            This is the date for when the strategy will begin. 
        end_date : dateTime
            This is the date for when the strategy will end. 
        pct_c : float64
            This value represents the percent chance for when to buy or sell shares. 
        shares : int64
            This value represents the number of shares to buy.
    
    Returns:
        Buy_in_out_portfolio : dataFrame
            This dataFrame contains the annualized returns and the sharpe ratio for each of the stocks that the user input into this function.
    """
    pct_changes = {}
    buy_ins = {}
    
    for e in etfs:
            pct_changes[e] = pct_c
            buy_ins[e] = shares

    Buy_in_out_portfolio = pd.DataFrame(columns=['Stock','Inception_date', 'Percent_change', 'Shares', 'Annualized_return', 'Sharpe_ratio'])

    for x in etfs:
        incep_date = stock_data_df.loc[stock_data_df['Stock'] == x, 'Inception_Date'].iloc[0]
        #print(x)
        #print(incep_date)
        portfolio_stats = Buy_Sell_Strategy([x], pct_changes, buy_ins, incep_date, end_date)

        Buy_in_out_portfolio.loc[len(Buy_in_out_portfolio.index)] = [x, incep_date, pct_changes[x], buy_ins[x], 
                                                                         portfolio_stats[x]['annualized_return'],
                                                                         portfolio_stats[x]['sharpe_ratio']]
    return Buy_in_out_portfolio
```

```python
def buy_sell_portfolio_multi(input_etfs, pct_c, shares, start_date, end_date):
    """
    This function preforms a buy-sell strategy for a portfolio of ETFs

    Args:
        input_etfs : list
            List of ETFs that will make up the portfolio.
        pct_c : float64
            Represents the percent change at which the function will buy or sell shares.
        shares : int64
            The number of shares that the function will start with.
        start_date : dateTime
            The date that represents when the strategy will begin.
        end_date : dateTime
            The date that represents when the strategy will end.

    Returns:
    Buy_in_out_portfolio : dataFrame
        Dataframe that contains the statistics for each of the ETFs.
    overall_df : dataFrame
        dataFrame that contains the overall statistics for the portfolio.
    """

    pct_changes = {}
    buy_ins = {}

    for e in input_etfs:
        #p = input(f"Enter percent change for {e}: ")
        #pct_changes.append(float(p))
        pct_changes[e] = pct_c
        #b = input(f"Enter amount of shares for {e}: ")
        #buy_ins.append(int(b))
        buy_ins[e] = shares

    #print(pct_changes)
    #print(buy_ins)


    # Calculate the portfolio statistics
    portfolio_stats = Buy_Sell_Strategy(input_etfs, pct_changes, buy_ins, start_date, end_date)
    #print(input_etfs)
    #print(pct_changes)
    #print(buy_ins)
    #print(" ")
    # Print the overall portfolio statistics
    print("Overall Portfolio Statistics:")
    for key, value in portfolio_stats['Combined'].items():
        print(f"{key}: {value:.2f}")

    # Print the individual stock statistics
    Buy_in_out_portfolio = pd.DataFrame(columns=['Stock', 'Percent_change', 'Shares', 'Total_investment', 'stock_shares_owned', 'Annualized_return', 'Sharpe_ratio'])

    print("\nIndividual Stock Statistics:")
    for stock_name, stats in portfolio_stats.items():
        if stock_name != 'Combined':
            print(f"{stock_name}:")
            for key, value in stats.items():
                print(f"  {key}: {value:.2f}")


    for x in input_etfs:
        Buy_in_out_portfolio.loc[len(Buy_in_out_portfolio.index)] = [x,pct_c, shares,  
                                                                     portfolio_stats[x]['total_investment'],
                                                                     portfolio_stats[x]['stock_shares_owned'],
                                                                     portfolio_stats[x]['annualized_return'],
                                                                     portfolio_stats[x]['sharpe_ratio']]
  
    combined_dict = portfolio_stats['Combined']
    overall_df = pd.DataFrame.from_dict(combined_dict, orient='index', columns=['Value'])
    overall_df.index.name = 'Overall Metrics'
    return Buy_in_out_portfolio, overall_df
```
